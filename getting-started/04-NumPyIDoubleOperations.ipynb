```python

import aadc
import numpy as np
import math
```

```python
s0 = 100
rfr = 0.01
sigma = 0.2
final_time = 1
num_tsteps = 100
SEED = 42
rng = np.random.default_rng(SEED)
dt = final_time/num_tsteps
num_paths = 1
```

```python
time = np.linspace(start=0., stop=final_time, num=num_tsteps)
time
```

```python
funcs = aadc.Functions()
funcs.start_recording()
```

```python
normal_samples_v = rng.standard_normal((num_paths, num_tsteps-1))
normal_samples_v
```

```python
normal_samples = np.array([aadc.idouble(x) for x in normal_samples_v.flatten()]).reshape((num_paths, num_tsteps-1))
normal_samples
```

```python
args = [x.mark_as_input() for x in normal_samples.flatten()]
args
```

```python
normal_samples*dt
```

```python
bm = np.concatenate((np.atleast_2d(np.zeros(num_paths)).T, np.cumsum(normal_samples*dt, axis=1)), axis=1)
bm
```

```python
np.cumsum(normal_samples*dt, axis=1)
```

```python
bm[0,0]
```

```python
bm[0,0] = bm[0,1] # without this I get error currently
bm[0,0]
```

```python
exp_arg = (rfr - sigma**2/2)*time + sigma*bm
exp_arg
```

```python
gbm = s0*np.exp((rfr - sigma**2/2)*time + sigma*bm)
gbm
```

```python
funcs.stop_recording()
```

```python
print(gbm)
```
